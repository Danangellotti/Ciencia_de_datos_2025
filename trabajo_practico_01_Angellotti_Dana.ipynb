## Ejercicio 01 **[2 puntos]**

Desarrollar una función `obtener_palabra` que me permita seleccionar un palabra de forma aleatorea de un repositorio de palabras de al menos 20 palabras, las cuales deben estar fijas en el código. No se puede utilizar el módulo random.

```python
# Lista fija de palabras para el juego del ahorcado
PALABRAS: list[str] = [
    "nirvana", "queen", "ledzeppelin", "pinkfloyd", "acdc",
    "metallica", "thebeatles", "aerosmith", "therollingstones", "u2",
    "gunsnroses", "radiohead", "pearljam", "bonjovi", "thewho",
    "thedoors", "deepurple", "ramones", "redhotchilipeppers", "black Sabbath"
]



def obtener_palabra(indice: int) -> str:
    """
    Devuelve una palabra del repositorio según el índice proporcionado.

    Args:
        indice (int): Índice entero utilizado para seleccionar la palabra.

    Returns:
        str: Palabra seleccionada del repositorio.

    Raises:
        ValueError: Si el índice está fuera del rango permitido.
    """
    total_palabras: int = len(PALABRAS)

    if indice < 0 or indice >= total_palabras:
        raise ValueError(f"Índice inválido. Debe estar entre 0 y {total_palabras - 1}.")

    return PALABRAS[indice]


# --- Simulación para elegir una palabra de forma "pseudoaleatoria" sin random ---

# Ingreso de datos controlado
entrada_usuario: str = input(f"Ingresá un número del 0 al {len(PALABRAS)-1}: ")

# Validación de entrada
if entrada_usuario.isdigit():
    indice_entero: int = int(entrada_usuario)
    try:
        palabra_seleccionada: str = obtener_palabra(indice_entero)
        print(f"\n La palabra seleccionada es: '{palabra_seleccionada}'")
    except ValueError as error:
        print(f" Error: {error}")
else:
    print(" Error: La entrada no es un número entero válido.")
```

## Ejercicio 02 **[2 puntos]**

Desarrollar una función `verificar_letra` que dada una letra verifique si se encuentra dentro de la palabra.

La verificación no se puede realizar con funciones predeterminadas de python como por ejemplo `index`, `in`, etc. Solo se puede utilizar el `=`.


```python
def verificar_letra(letra: str, palabra: str) -> bool:
    """
    Verifica si una letra está dentro de la palabra, comparando carácter por carácter.

    Args:
        letra (str): Letra a buscar.
        palabra (str): Palabra donde buscar la letra.

    Returns:
        bool: True si la letra está, False en caso contrario.
    """
    posicion: int = 0
    encontrada: bool = False

    while posicion < len(palabra):
        if palabra[posicion] == letra:
            encontrada = True
        posicion = posicion + 1

    return encontrada


#  Ejemplo de prueba

palabra_prueba: str = "queen"
letra_prueba: str = input("Ingresá una letra a verificar: ")

# Validamos que solo se ingrese una letra
if len(letra_prueba) != 1:
    print(" Error: Ingresá solo una letra.")
else:
    resultado: bool = verificar_letra(letra_prueba, palabra_prueba)
    if resultado:
        print(f" La letra '{letra_prueba}' está en la palabra '{palabra_prueba}'.")
    else:
        print(f" La letra '{letra_prueba}' NO está en la palabra '{palabra_prueba}'.")
```

## Ejercicio 03 **[2 puntos]**

Desarrollar una función `completar_letras_ocultas` que complete todas las posiciones donde se encuentra la letra ingresada. En caso de que una letra se repita, la debo completra por cada aparición.


```python
def completar_letras_ocultas(palabra: str, letra: str, progreso_actual: list[str]) -> list[str]:
    """
    Reemplaza guiones bajos por la letra adivinada en todas sus posiciones correctas.

    Args:
        palabra (str): Palabra a adivinar.
        letra (str): Letra ingresada por el usuario.
        progreso_actual (list[str]): Lista con el estado actual de las letras adivinadas.

    Returns:
        list[str]: Lista actualizada con la letra en sus posiciones correspondientes.
    """
    indice: int = 0
    while indice < len(palabra):
        if palabra[indice] == letra:
            progreso_actual[indice] = letra
        indice = indice + 1

    return progreso_actual


#  Prueba del funcionamiento

# Ejemplo: palabra a adivinar
palabra_a_adivinar: str = "queen"

# Estado inicial (todo oculto con "_")
progreso_juego: list[str] = ["_"] * len(palabra_a_adivinar)

# Simulación: el usuario adivinó la letra 'a'
letra_ingresada: str = input("Ingresá una letra para completar la palabra: ")

if len(letra_ingresada) == 1:
    progreso_actualizado: list[str] = completar_letras_ocultas(palabra_a_adivinar, letra_ingresada, progreso_juego)
    print(f" Progreso actual: {' '.join(progreso_actualizado)}")
else:
    print(" Error: Solo se puede ingresar una letra.")
```

## Ejercicio 04 **[2 puntos]**

Desarrollar una función `errores_cometidos` que administre los errores que se van cometiendo.

Los errores generan los siguientes gráficos.

```python
error_1:str = """

   -----
   |
   |
   |
   |
  ---
"""
error_2:str = """

   -----
   |   O
   |
   |
   |
  ---
"""
error_3:str = """

   -----
   |   O
   |   |
   |
   |
  ---
"""
error_4:str = """

   -----
   |   O
   |  /|
   |
   |
  ---
"""
error_5:str = """

   -----
   |   O
   |  /|\\
   |
   |
  ---
"""
error_6:str = """

   -----
   |   O
   |  /|\\
   |  /
   |
  ---
"""
error_7:str = """

   -----
   |   O
   |  /|\\
   |  / \\
   |
  ---
"""
```

```python
def errores_cometidos(contador_errores: int) -> str:
    """
    Devuelve el gráfico del ahorcado según la cantidad de errores cometidos.

    Args:
        contador_errores (int): Número de errores cometidos.

    Returns:
        str: Gráfico del estado actual del ahorcado.

    Raises:
        ValueError: Si el número de errores no está entre 0 y 7.
    """
    if contador_errores < 0 or contador_errores > 7:
        raise ValueError("El número de errores debe estar entre 0 y 7.")

    error_0: str = """

       -----
       |
       |
       |
       |
      ---
    """
    error_1: str = """

       -----
       |   O
       |
       |
       |
      ---
    """
    error_2: str = """

       -----
       |   O
       |   |
       |
       |
      ---
    """
    error_3: str = """

       -----
       |   O
       |  /|
       |
       |
      ---
    """
    error_4: str = """

       -----
       |   O
       |  /|\\
       |
       |
      ---
    """
    error_5: str = """

       -----
       |   O
       |  /|\\
       |  /
       |
      ---
    """
    error_6: str = """

       -----
       |   O
       |  /|\\
       |  / \\
       |
      ---
    """

    graficos: list[str] = [
        error_0, error_1, error_2, error_3, error_4, error_5, error_6
    ]

    return graficos[contador_errores]


#  Prueba de ejemplo
errores: int = int(input("Ingresá la cantidad de errores cometidos (0 a 6): "))

try:
    grafico: str = errores_cometidos(errores)
    print(grafico)
except ValueError as error:
    print(f" Error: {error}")
```

## Punto 05 **[2 puntos]**

Desarrollar el programa que utilice las funciones desarrolladas anteriormente y que me permita jugar al juego.

El juego debe permitir adivinar varias palabras dentro de la misma ejecución, es decir, debe mostrar algún cartel para continuar jugando. En caso de continuar jugando no se puede adivinar las misma palabra 2 veces consecutivas.

```python
def jugar_ahorcado() -> None:
    """
    Ejecuta el juego del Ahorcado utilizando las funciones desarrolladas.
    Permite jugar múltiples rondas sin repetir la palabra inmediatamente anterior.
    """
    ultima_palabra_usada: str = ""
    seguir_jugando: bool = True

    while seguir_jugando:
        # Seleccionar palabra sin repetir la anterior
        palabra_actual: str = ""
        while palabra_actual == "" or palabra_actual == ultima_palabra_usada:
            entrada: str = input(f"Ingresá un número del 0 al {len(PALABRAS) - 1} para seleccionar la palabra: ")
            if entrada.isdigit():
                indice: int = int(entrada)
                try:
                    palabra_actual = obtener_palabra(indice)
                except ValueError as error:
                    print(f" Error: {error}")
            else:
                print(" Error: Solo se permiten números.")

        ultima_palabra_usada = palabra_actual
        progreso: list[str] = ["_"] * len(palabra_actual)
        letras_erroneas: list[str] = []
        errores: int = 0
        letras_adivinadas: list[str] = []

        print("\n ¡Comienza el juego del Ahorcado!")
        print(f" La palabra tiene {len(palabra_actual)} letras.")

        while "_" in progreso and errores < 7:
            print("\n Progreso:", " ".join(progreso))
            print(" Letras incorrectas:", ", ".join(letras_erroneas))
            print(errores_cometidos(errores))

            letra: str = input(" Ingresá una letra : ")

            if len(letra) != 1:
                print(" Por favor ingresá solo una letra.")
                continue

            # Verificamos si ya se adivinó o erró esa letra
            letra_repetida: bool = False
            for l in letras_adivinadas:
                if l == letra:
                    letra_repetida = True
            for l in letras_erroneas:
                if l == letra:
                    letra_repetida = True

            if letra_repetida:
                print(" Ya usaste esa letra. Probá con otra.")
                continue

            if verificar_letra(letra, palabra_actual):
                letras_adivinadas.append(letra)
                progreso = completar_letras_ocultas(palabra_actual, letra, progreso)
            else:
                letras_erroneas.append(letra)
                errores = errores + 1

        # Fin de la partida
        if "_" not in progreso:
            print("\n ¡Ganaste! La palabra era:", palabra_actual)
        else:
            print(errores_cometidos(errores))
            print(" ¡Perdiste! La palabra era:", palabra_actual)

        # Preguntar si quiere seguir
        respuesta: str = input("\n¿Querés jugar otra ronda? (s/n): ")
        if respuesta.lower() != "s":
            seguir_jugando = False
            print(" ¡Gracias por jugar!")

# Ejecutar el juego
jugar_ahorcado()
```
